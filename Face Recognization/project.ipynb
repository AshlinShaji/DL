```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import cmake
import face_recognition as fr
import cv2 
```

```python
path='images'
```

```python
os.listdir(path)
```

```python
mylist=os.listdir(path)
mylist
```

```python
imgs=[]
classnames=[]
for i in mylist:
    imgpath=os.path.join(path,i)
    current_img=cv2.imread(imgpath)
    print(current_img)
    imgs.append(current_img)
    classnames.append(i.split(".")[0])
    
```

```python
#face location and face encoding
def faceencodings(images):
    encodelist=[]
    for img in images:
        img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        face_in_frame=fr.face_locations(img)
        print(face_in_frame)
        encoding=fr.face_encodings(img,face_in_frame)[0]#0 to convert to 1d
        encodelist.append(encoding)
    return encodelist

```

```python
encode_known_face=faceencodings(imgs)
encode_known_face
```

```python
video=cv2.VideoCapture(0)
while True:
    suc,img=video.read()
    img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    face_in_frame=fr.face_locations(img1)
    # print(face_in_frame)
    encoded_face=fr.face_encodings(img1,face_in_frame)
    # print(encoded_face)
    
    for face_loc,enc in zip(face_in_frame,encoded_face):
        matches=fr.compare_faces(encode_known_face,enc)
        face_distance=fr.face_distance(encode_known_face,enc)
        math_index=np.argmin(face_distance)
        if matches[math_index]==True:
            name=classnames[math_index]
            print(name)
            y1,x2,y2,x1=face_loc
            cv2.rectangle(img,(x1,y1),(x2,y2),(0,255,0),2)
            cv2.putText(img,name,(x1,y1-10),cv2.FONT_HERSHEY_SIMPLEX,0.9,(255,0,0),2) 
        


    cv2.imshow('video',img)
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break

video.release()
cv2.destroyAllWindows()
```

```python
video = cv2.VideoCapture(0)

while True:
    suc, img = video.read()
    if not suc:
        print("Failed to capture video")
        break

    img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    face_in_frame = fr.face_locations(img1)
    encoded_face = fr.face_encodings(img1, face_in_frame)

    for face_loc, enc in zip(face_in_frame, encoded_face):
        matches = fr.compare_faces(encode_known_face, enc)
        face_distance = fr.face_distance(encode_known_face, enc)
        match_index = np.argmin(face_distance)

        y1, x2, y2, x1 = face_loc

        if matches[match_index]:
            name = classnames[match_index]
            print(name)
        else:
            name = "Unknown"
            color = (0, 0, 255)  # Red
            print("Unknown Face")

        cv2.rectangle(img, (x1, y1), (x2, y2), color, 2)
        cv2.putText(img, name, (x1, y1 - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.9, color, 2)

    cv2.imshow('video', img)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

video.release()
cv2.destroyAllWindows()
```
