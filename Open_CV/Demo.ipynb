```python
#image convert to binary image
#we cannot convert image directly to binary so we first convrt image to greyscale then binary
#threshold 2 variable
thres,bin_img=cv2.threshold(img_gray,125,255,cv2.THRESH_BINARY)
cv2.imshow('Binary_image',bin_img)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
thres
```

```python
cv2.imwrite('baby1_bin_img.jpg',bin_img)
```

```python
#image display is used matplotlib

img1=cv2.imread('baby1.jpg')
img1
```

```python
import matplotlib.pyplot as plt
plt.imshow(img1)
```

```python
#to convert bgr image to rgb
conv_img=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img)
```

```python
#Bounding Box
#Rectangle
#left top , Right Bottom
#(180,100)  (350,275)
cv2.rectangle(conv_img,pt1=(180,100),pt2=(350,275),color=(255,0,0),thickness=3)
plt.imshow(conv_img)
```
