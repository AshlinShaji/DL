```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('rajani.jpeg')
img
```

```python
plt.imshow(img)
```

```python
conv_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img)
```

```python
face_data=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
face_data
```

```python
#face detect
faces=face_data.detectMultiScale(conv_img)
faces
```

```python
for (x,y,w,h) in faces:
    cv2.rectangle(conv_img,(x,y),(x+w,y+h),(255,0,0),2)
```
