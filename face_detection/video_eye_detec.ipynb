```python
import cv2
```

```python
eye_cas=cv2.CascadeClassifier("haarcascade_eye.xml")
eye_cas
```

```python
#liv video
Video=cv2.VideoCapture(0)

#READ
while True:
    SUCCESS, img=Video.read()
    grey_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    eyes=eye_cas.detectMultiScale(grey_img,minNeighbors=5)
    for x,y,w,h in eyes:    
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    
    cv2.imshow('Webcam',img)
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break

Video.release()
cv2.destroyAllWindows()
```

```python
#video uploaded

video=cv2.VideoCapture('eye_video.mp4')

#read
while True:
    SUCCESS, img=video.read()
    if SUCCESS:
        grey_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        eyes=eye_cas.detectMultiScale(grey_img,minNeighbors=10)
        for x,y,w,h in eyes:    
            cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
        
        img = cv2.resize(img, (800, 600))
        cv2.imshow('Video',img)
        if cv2.waitKey(1) & 0xFF==ord('q'):
            break
    else:
        print("Video ended")
        break
video.release()
cv2.destroyAllWindows()    
```
