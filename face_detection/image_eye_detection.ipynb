```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("fafa.jpg")
img
```

```python
plt.imshow(img)
```

```python
conv_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img)
```

```python
eye_data=cv2.CascadeClassifier("haarcascade_eye.xml")
eyes=eye_data.detectMultiScale(conv_img)
eyes
```

```python
for x,y,w,h in eyes:
    cv2.rectangle(conv_img,(x,y),(x+w,y+h),(255,0,0),2)
```

```python
plt.imshow(conv_img)
```
