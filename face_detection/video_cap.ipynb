```python
import cv2
```

```python
face_cas=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
face_cas
```

```python
#Live video capturing
video=cv2.VideoCapture(0)
#read 
while True:
    sucesss,img=video.read()
    grey_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces=face_cas.detectMultiScale(grey_img,minNeighbors=5)
    for (x,y,w,h) in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),3)

    cv2.imshow("Webcam",img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

video.release()
cv2.destroyAllWindows()
```

```python
#video uploaded
video=cv2.VideoCapture('random_vid.mp4')
#read 
while True:
    sucesss,img=video.read()
    if sucesss:

        grey_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        faces=face_cas.detectMultiScale(grey_img,minNeighbors=10)
        for (x,y,w,h) in faces:
            cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),3)

        img = cv2.resize(img, (800, 600))
        cv2.imshow("Webcam",img)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        print("Video ended")
        break

video.release()
cv2.destroyAllWindows()
```
