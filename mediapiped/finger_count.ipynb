```python
import mediapipe as mp
import cv2
```

```python
mphands=mp.solutions.hands
mpdraw=mp.solutions.drawing_utils
hand=mphands.Hands()
```

```python
video=cv2.VideoCapture(0)
while True:
    suc,img=video.read()
    img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    result=hand.process(img1)
    # print(result.multi_hand_landmarks)
    tipid=[4,8,12,16,20]
    lmlist=[]

    if result.multi_hand_landmarks:
        for x in result.multi_hand_landmarks:
            for id,lm in enumerate(x.landmark):
                cx=lm.x
                cy=lm.y
                lmlist.append([id,cx,cy])
                # print(lmlist)
                if len(lmlist)!=0 and len(lmlist)==21:
                    fingercount=[]
                    #thump
                    if lmlist[12][1]<lmlist[20][1]:
                        if lmlist[4][1]>lmlist[3][1]:
                            fingercount.append(0)
                        else:
                            fingercount.append(1)
                    else:
                        if lmlist[4][1]<lmlist[3][1]:
                            fingercount.append(0)
                        else:
                            fingercount.append(1) 

                    for i in range(1,5):
                        if lmlist[tipid[i]][2]>lmlist[tipid[i]-2][2]:
                            fingercount.append(0)
                        else:
                            fingercount.append(1) 
                    # print(fingercount)   
                    if len(fingercount)!=0:
                        totalfingers=fingercount.count(1)
                    print(totalfingers)
                    cv2.putText(img,"Count:"+str(totalfingers),(50,150),cv2.FONT_HERSHEY_SIMPLEX,2,(255,0,0),5)
            mpdraw.draw_landmarks(img,x,mphands.HAND_CONNECTIONS,mpdraw.DrawingSpec(color=(0,0,255),circle_radius=5,thickness=2),mpdraw.DrawingSpec(color=(0,255,0),circle_radius=5,thickness=2))

    cv2.imshow("Image",img)
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break

video.release()
cv2.destroyAllWindows()
```
