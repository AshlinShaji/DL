```python
df.drop(['Id'],axis=1,inplace=True)
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.30,random_state=1)
x_train
```

```python
x_test
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)  # if you have test data
```

```python
x_test
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
ann=Sequential()
```

```python
ann.add(Dense(units=4,activation='relu'))
ann.add(Dense(units=10,activation='relu'))
ann.add(Dense(units=6,activation='relu'))
ann.add(Dense(units=3,activation='softmax'))
```

```python
ann.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
ann.fit(x_train,y_train,epochs=100,batch_size=10)
```

```python

ann.evaluate(x_test,y_test)
```

```python
ann.summary()
```

```python
y_pred=ann.predict(scaler.transform([[2.3,1,5,3]]))
y_pred
```

```python
ind=y_pred.argmax()
ind
```

```python
pred=le.inverse_transform([ind])
pred
```
