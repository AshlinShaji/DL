```python
import cv2
import re
import pytesseract
```

```python

img=cv2.imread('resume.jpg')
img
```

```python
cv2.imshow('Resume Image',img)
cv2.waitKey(5000)  
cv2.destroyAllWindows()
```

```python
pytesseract.pytesseract.tesseract_cmd = r'C:\Users\ashli\Documents\tesseract ocr\tesseract.exe'
```

```python
text=pytesseract.image_to_string(img)
print(text)
```

```python
data=pytesseract.image_to_data(img,output_type=pytesseract.Output.DICT)
data
```

```python
n=len(data['text'])
n
```

```python
for i in range(n):
    if re.search(r"[A-Za-z0-9._%+-]+\s*@\s*[A-Za-z0-9.-]+\s*\.\s*[A-Za-z]{2,}",data['text'][i]):
        print(data['text'][i])
        x=int(data['left'][i])
        y=int(data['top'][i])
        w=int(data['width'][i])
        h=int(data['height'][i])
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
cv2.imshow('Resume Image',img)
cv2.waitKey(5000)   
cv2.destroyAllWindows()                
```
